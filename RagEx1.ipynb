```python
from langchain_community.document_loaders import UnstructuredFileIOLoader

with open('Python%20Built-In%20Functions.pdf', 'rb') as f:
    loader = UnstructuredFileIOLoader(file=f)
    documents = loader.load()
```

```python
pip install langchain_text_splitters
```

```python
from langchain_text_splitters import CharacterTextSplitter
text_splitter = CharacterTextSplitter(chunk_size=100,chunk_overlap=50)
```

```python
text_splitter
```

```python
texts= text_splitter.split_documents(documents)
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
persist_directory = "vector_db"
```

```python
from langchain_chroma import Chroma

vector_db = Chroma.from_documents(documents=texts,embedding=embeddings,persist_directory=persist_directory)
```

```python
print(vector_db)
```

```python
retriever = vector_db.as_retriever()
```

```python
from langchain_groq import ChatGroq
llm = ChatGroq(model="llama-3.3-70b-versatile",temperature=0.2)
```

```python
from langchain_classic.chains import RetrievalQA

qa_chain = RetrievalQA.from_chain_type(
           llm=llm,
           chain_type="stuff",
           retriever=retriever,
           return_source_documents=True
           )
```

```python
query = 'what are the function'
qa_result = qa_chain.invoke({"query":query})
```

```python
print(qa_result)
```
